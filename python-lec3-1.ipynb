グラフに情報を付け足そう。<br>
・タイトル：basic plot<br>
・x軸ラベル：x axis<br>
・y軸のラベル：y axis<br>

```python
plt.figure(figsize=(3,3))
plt.plot([0,2,4,6,8],[0,6,12,24,48], label='curve')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xlim(2,8)
plt.title("basic plot")
plt.legend(loc='upper left')
```

Numpy arrayと組み合わせる。numpyをインポートしよう。（呼び出し名はnpとする）

```python
import numpy as np
```

x=0~10(両端含む)の間で21点とるnumpy arrayとし、printで中身を出力<br>
y=x^2<br>
z=(x-1)^3<br>
w={(x-2)(x-4)}^2<br>
とする。

```python
x=np.linspace(0,10,21)
print(x)
y=x**2
z=(x-1)**3
w=((x-2)*(x-4))**2
```

pyplotインターフェースでy,z,wを重ねて描画。凡例を忘れずに。

```python
plt.plot(x,y,label="y=x^2")
plt.plot(x,z,label="z=(x-1)^3")
plt.plot(x,w,label="w=((x-2)(x-4))^2")
plt.legend()
```

この３つの曲線を別々の図にプロットして並べたい場合、subplotを用いる。<br>
yのグラフ、zのグラフ、wのグラフを横一列に並べて描画してみよう。<br>
隣のグラフに数字が重ならないようにしよう。

```python
plt.subplot(1,3,1) 
plt.plot(x,y,label="y=x^2")
plt.legend()

plt.subplot(1,3,2)
plt.plot(x,z,label="z=(x-1)^3")
plt.legend()

plt.subplot(1,3,3)
plt.plot(x,w,label="w=((x-2)(x-4))^2")
plt.legend()
```

```python
#やたら縦長になった理由はfigureのdefalt sizeが(8,6)、つまり横に8 inch、縦に6 inchであるから
#重なりが生じて嫌な感じなので、tight layout を追記
```

```python
plt.subplot(1,3,1) #中の引数は●行●列●番目と覚える
plt.plot(x,y,label="y=x^2")
plt.legend()

plt.subplot(1,3,2)
plt.plot(x,z,label="z=(x-1)^3")
plt.legend()

plt.subplot(1,3,3)
plt.plot(x,w,label="w=((x-2)(x-4))^2")
plt.legend()

plt.tight_layout()
```

<b>オブジェクト指向インターフェース</b>

空のfigureを生成し、figという変数に代入。
axというaxesを追加し、その中にy,z,wを全てプロットし、凡例をつける。
グラフのサイズは(3,3)とする。

```python
fig = plt.figure(figsize=(3,3))
ax=fig.add_subplot(1,1,1)
ax.plot(x,y,label="y=x^2")
ax.plot(x,z,label="z=(x-1)^3")
ax.plot(x,w,label="w=((x-2)(x-4))^2")
ax.set_xlim(0,10)
ax.legend()
```

空のfigureをfigとして生成する。<br>
figの全面にaxというaxesを生成する。<br>
figの左から20%、下から30%の位置に存在する点Aと、figの中心の点Bを対角線とするようなaxesを生成しax2とする。<br>
figの左から60%、下から60%の位置に存在する点Aと、figの中心の点Bを対角線とするようなaxesを生成しax3とする。<br>

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.3,0.3,0.2])
ax3=fig.add_axes([0.6,0.6,0.4,0.4])
```

先ほどpyplotインタフースで描画した3つのグラフを、<br>
オブジェクト指向インターフェースを用いて、重ねて表示しよう。<br>

```python
fig,ax= plt.subplots(1,1,figsize=(3,3))
ax.plot(x,y,label="y=x^2")
ax.plot(x,z,label="z=(x-1)^3")
ax.plot(x,w,label="w=((x-2)(x-4))^2")
ax.legend()
ax.set_title("three curves")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
```

```python
ax
```

上の３つの曲線を別々のグラフにプロットし、3つのグラフを横一列に並べよう。<br>
plt.subplots()を用いること。<br>

```python
fig,ax = plt.subplots(1,3,figsize=(12,4))    #●行●列を定義。グラフ全体の話をしているので「●番目」は必要ない。
print(ax)#axはarray形式であることを確認
ax[0].plot(x,y,label="y=x^2")
ax[1].plot(x,z,label="z=(x-1)^3")
ax[2].plot(x,w,label="w=((x-2)(x-4))^2")
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x axis")
    ax[i].set_ylabel("y axis")
fig.tight_layout()
```

先ほど横一列に並べた3つのグラフを今度は<br>
図のように並べてみよう。<br>
これもplt.subplots()を用いること。<br>

```python
fig,ax = plt.subplots(2,2)  #●行●列(axesの配置)を定義。グラフ全体の話をしているので「●番目」は必要ない。
ax[0,0].plot(x,y,label="y=x^2")
ax[0,0].legend()
ax[0,1].plot(x,z,label="z=(x-1)^3")
ax[0,1].legend()
ax[1,0].plot(x,w,label="w=((x-2)(x-4))^2")
ax[1,0].legend()
```

定義域によって異なる関数をプロットする場合

```python
x=np.linspace(0,10,200)
condition=[x<3,(x>=3)&(x<=7),x>7]
function=[lambda x:x, lambda x: 2*x-3, lambda x :18-x]
y=np.piecewise(x,condition,function)
plt.plot(x, y)
```
